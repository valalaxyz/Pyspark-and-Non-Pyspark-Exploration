Implementasi model klasifikasi non pyspark

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['trustLevel',	'totalScanTimeInSeconds',	'grandTotal',	'lineItemVoids',	'scansWithoutRegistration',	'quantityModifications',	'scannedLineItemsPerSecond',	'valuePerSecond',	'lineItemVoidsPerPosition']
df[columns_to_scale] = standardScaler.fit_transform(df[columns_to_scale])
df.head()
```

```python
#Pemisahan Variabel Prediktor (X) dan Variabel Target (Y)
X1=df.iloc[:,0:8].values
Y1=df.iloc[:,9].values

#Pemisahan data Training dan Testing dengan testing 20% dan training 80%
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X1,Y1, test_size = 0.2)
```

```python
#Klasifikasi dengan Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import time

#training data
start = time.time()

dtc = DecisionTreeClassifier()
dtc.fit(X_train1, Y_train1)
print('Waktu training:', (time.time() - start), 'detik')
```

```python
#testing data
start = time.time()
Y_pred11 = dtc.predict(X_test1)

print('Accuracy = ', accuracy_score(Y_test1, Y_pred11))
print('Waktu uji akurasi:', (time.time() - start), 'detik')

print("=" *50)
print(classification_report(Y_test1,Y_pred11))
```

```python
# Visualisasi korelasi

# Menggabungkan semua fitur numerik menjadi satu kolom vektor
assembler = VectorAssembler(inputCols=df.columns, outputCol="features")
df_assembled = assembler.transform(df)

# Menghitung matriks korelasi menggunakan metode corr
correlation_matrix = Correlation.corr(df_assembled, "features").collect()[0][0]

# Konversi matriks korelasi menjadi Pandas DataFrame

correlation_pd = pd.DataFrame(correlation_matrix.toArray(), columns=df.columns, index=df.columns)

# Membuat heatmap menggunakan Seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_pd, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title("Correlation Matrix dalam Pysaprk")
plt.show()
```

```python
#set kolom menjadi satu array bernama features
columns_to_input = ['trustLevel',	'totalScanTimeInSeconds',	'grandTotal',	'lineItemVoids',	'scansWithoutRegistration',	'quantityModifications',	'scannedLineItemsPerSecond',	'valuePerSecond',	'lineItemVoidsPerPosition']
vector_assembler = VectorAssembler(inputCols=columns_to_input,outputCol="Features")
df_temp = vector_assembler.transform(df)
df_temp.show(10)

#memilih kolom output (fraud dan features)
df = df_temp.drop('trustLevel',	'totalScanTimeInSeconds',	'grandTotal',	'lineItemVoids',	'scansWithoutRegistration',	'quantityModifications',	'scannedLineItemsPerSecond',	'valuePerSecond',	'lineItemVoidsPerPosition')
df.show(20)
# set data training 80% dan testing 20%
(training,testing) = df.randomSplit([0.8,0.2])
```

Menampilkan data pertama adalah data fitur selain fraud ditambah dengann kolom features yang berisikan array fitur selain fraud. Lalu untuk konsol kedua menampilkan nilai features dengan fraudnya

Decision Tree/Random Forest

```python
import time
from pyspark.ml.classification import DecisionTreeClassifier

#Menyimpan waktu awal
start = time.time()

#Training model Decision Tree
dt = DecisionTreeClassifier(labelCol="fraud", featuresCol="Features")
model = dt.fit(training)

#Cetak waktu Training
print('Waktu training:', (time.time() - start), 'detik')
```

```python
model
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.sql.types import FloatType
import pyspark.sql.functions as F

# testing
start = time.time()

# menerapkan model yang sudah dilatih untuk mendapatkan prediksi
predictions = model.transform(testing)

# sep kolom prediction dan fraud
pred = predictions.select("prediction", "fraud")

evaluator = MulticlassClassificationEvaluator(labelCol="fraud", predictionCol="prediction",metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Model DecisionTreeClassifier")
print('Waktu testing:', (time.time() - start), 'detik')
pred.show(10)
print("Tingkat Error = %g " % (1.0 - accuracy))
print("Nilai Akurasi= %g " % accuracy)
```

Menampilkan nilai akurasi untuk model klasifikasi Decision Tree

Naive Bayes

```python
from pyspark.ml.classification import NaiveBayes
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Menyimpan waktu awal
start = time.time()

# Training model Naive Bayes
nb = NaiveBayes(labelCol="fraud", featuresCol="Features")
model = nb.fit(training)

# Cetak waktu Training
print('Waktu training:', (time.time() - start), 'detik')
```

```python
# Testing
start = time.time()

# Menerapkan model yang sudah dilatih untuk mendapatkan prediksi
predictions = model.transform(testing)

# Evaluasi menggunakan MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(labelCol="fraud", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Model NaiveBayes")
print('Waktu testing:', (time.time() - start), 'detik')
pred.show(10)
print("Tingkat Error = %g " % (1.0 - accuracy))
print("Nilai Akurasi = %g " % accuracy)
```

Menampilkan nilai akurasi untuk model klasifikasi NaiveBayes


SVMs

```python
from pyspark.ml.classification import LinearSVC
# Menyimpan waktu awal
start = time.time()

# Training model Support Vector Machines
svm = LinearSVC(labelCol="fraud", featuresCol="Features")
model = svm.fit(training)

# Cetak waktu Training
print('Waktu training:', (time.time() - start), 'detik')
```

```python
# Testing
start = time.time()

# Menerapkan model yang sudah dilatih untuk mendapatkan prediksi
predictions = model.transform(testing)

# Evaluasi model
evaluator = MulticlassClassificationEvaluator(labelCol="fraud", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)

print("Model Support Vector Machines (SVMs)")
print('Waktu testing:', (time.time() - start), 'detik')
predictions.select("prediction", "fraud").show(10)
print("Tingkat Error = %g " % (1.0 - accuracy))
print("Nilai Akurasi= %g " % accuracy)
```

Menampilkan model klasifikasi Model Support Vector Machines (SVMs)

Logistic Regression

```python
from pyspark.ml.classification import LogisticRegression

# Menyimpan waktu awal
start = time.time()

# Training model Logistic Regression
lr = LogisticRegression(labelCol="fraud", featuresCol="Features", maxIter=10, regParam=0.3, elasticNetParam=0.8)
model = lr.fit(training)

# Cetak waktu Training
print('Waktu training:', (time.time() - start), 'detik')
```

```python
# Testing
start = time.time()

# Menerapkan model yang sudah dilatih untuk mendapatkan prediksi
predictions = model.transform(testing)

# Evaluasi model
evaluator = MulticlassClassificationEvaluator(labelCol="fraud", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)

print("Model Logistic Regression")
print('Waktu testing:', (time.time() - start), 'detik')
predictions.select("prediction", "fraud").show(10)
print("Tingkat Error = %g " % (1.0 - accuracy))
print("Nilai Akurasi= %g " % accuracy)
```
